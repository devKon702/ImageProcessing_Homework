```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr=np.array(list(float(y) for y in range(0,8)))
[COLUMN ,ROW] = np.meshgrid(arr,arr)
u0 = 2.0
v0 = 2.0
```

Part 1

```python
def create_imaginary(input):
    return np.round(np.imag(input) * (10**4)) * (10**(-4))
def create_real(input):
    return np.round(np.real(input) * (10**4)) * (10**(-4))
def create_center(input):
    return np.fft.fftshift(np.fft.fft2(input))
def create_full_contrast(input):
    min_val = np.min(input)
    max_val = np.max(input)
    if(min_val == max_val):
        return np.zeros(input.shape)
    scale = 255/(max_val - min_val)
    return np.round((input - min_val) * scale)
def format_print(name, data):
    print(f'{name}: ')
    print(data)
def show2img(img1, title1, img2, title2):
    plt.figure(figsize=(10, 10))
    plt.subplot(1, 2, 1)
    plt.title(title1)
    plt.imshow(img1,cmap='gray')

    plt.subplot(1, 2, 2)
    plt.title(title2)
    plt.imshow(img2,cmap='gray')
```

```python
def create_I1(m,n):
    I1=0.5 * np.exp(2j*np.pi/8*(u0*m + v0*n))
    return I1
I1=create_I1(COLUMN,ROW)
```

```python
I1_real = create_real(I1)
I1_imag = create_imaginary(I1)
format_print("I1 Real", I1_real)
format_print("I1 Imaginary", I1_imag)
show2img(create_full_contrast(I1_real),"I1 Real",create_full_contrast(I1_imag),"I1 Imaginary")
```

```python
Itilde1 = create_center(I1)
Itilde1_real = create_real(Itilde1)
Itilde1_imag = create_imaginary(Itilde1)
format_print("Re[DFT(I1)]", Itilde1_real)
format_print("Im[DFT(I1)]", Itilde1_imag)
```

Part 6

```python
def get_img(path):
    return np.fromfile(path,dtype=np.uint8).reshape(256,256)
def show_multi_image(img):
    plt.figure(figsize=(14, 10))
    plt.subplot(1, 5, 1)
    plt.title("Origin")
    plt.imshow(create_full_contrast(img),cmap='gray')

    center_img = create_center(img)
    real_img = create_real(center_img)
    imag_img = create_imaginary(center_img)
    log_magnitude_spectrum = np.log(np.abs(center_img)+1)
    phase_spectrum = np.angle(center_img)

    plt.subplot(1, 5, 2)
    plt.title("Real")
    plt.imshow(create_full_contrast(real_img),cmap='gray')

    plt.subplot(1, 5, 3)
    plt.title("Imaginary")
    plt.imshow(create_full_contrast(imag_img),cmap='gray')

    plt.subplot(1, 5, 4)
    plt.title("Log-magnitude")
    plt.imshow(create_full_contrast(log_magnitude_spectrum),cmap='gray')

    plt.subplot(1, 5, 5)
    plt.title("Phase")
    plt.imshow(create_full_contrast(phase_spectrum),cmap='gray')
```

```python
cam_img = get_img("camera.bin")
head_img = get_img("head.bin")
eyeR_img = get_img("eyeR.bin")
saleman_img = get_img("salesman.bin")
```

```python
show_multi_image(cam_img)
show_multi_image(head_img)
show_multi_image(eyeR_img)
show_multi_image(saleman_img)
```

Part 7

```python
I6 = get_img("camera.bin")
DFT_I6 = np.fft.fft2(I6)

mag = np.abs(DFT_I6)
phase = np.angle(DFT_I6)

J1 = create_real(np.fft.ifft2(mag))
J2 = create_real(np.fft.ifft2(np.exp(1j * phase)))
JJ1 = np.log(J1)

show2img(create_full_contrast(JJ1), "JJ1", create_full_contrast(J2), "J2")
```
